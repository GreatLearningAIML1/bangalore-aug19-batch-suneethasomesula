```python
stop_words = set(stopwords.words('english')) #Remove all the English Stop words
```

```python
import re, string
# Function to Remove unwanteed Charactersand spaces Before Vectorization
def rem_uchar(string): 
   try:    
    string = re.sub(r'^https?:\/\/<>.*[\r\n]*', '', string, flags=re.MULTILINE)
    string = re.sub(r"[^A-Za-z]", " ", string)         
    words = string.strip().lower().split()    
    #words = [w for w in words if len(w)>=1]
    words = [w for w in words if not w in stop_words]
    return " ".join(words)	
   except:
    return ""
```

```python
#Function to convert all text to lower case before vectorization
def case_lower(string):
  try:
    string = re.sub(r"[^A-Za-z]", "", string)         
    words = string.strip().lower().split()    
    #words = [w for w in words if len(w)>=1]
    words = [w for w in words if not w in stop_words]
    return " ".join(words)	
  except:
    return ""
```

```python
import numpy as np
import pandas as pd
Bank_df=pd.read_csv('bank.csv')
```

```python
Bank_df.head()
```

```python
Bank_df.isnull().sum()
```

```python
drop_list=["CustomerId","Exited","Surname"]
New_Bank_df=Bank_df.drop(drop_list,axis=1)
```

```python
New_Bank_df.head()
Bank_df.head()
```

```python
New_Bank_df.groupby(["Geography"]).count()
```

```python
#Encoding all the object data types.
New_Bank_obj = New_Bank_df.select_dtypes(include=['object']).copy()
cleanup_nums = {"Gender": {"Female": 0, "Male": 1 },"Geography":{"France": 0, "Germany": 1,"Spain":2 }}
X_df=New_Bank_df.replace(cleanup_nums, inplace=False)# Replacing the categorical Target variable with the binary encoded values.
X_df.head()
```

```python
X=X_df.values
#print("The following are the features\n",col_names)
Y_df=Bank_df.iloc[:,13]
Y=Y_df.values
Y_df.head()
```

```python
print(X_df.shape,Y_df.shape)
print("Target Set is \n",Y_df.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=7)
```

```python
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
import tensorflow as tf #Initialize & build the model
print(tf.__version__)
```

```python
tf.random.set_seed(42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Initialize Sequential model
model = Sequential()
#Add input layer
model.add(Dense(11, input_dim = 11, kernel_initializer = 'uniform', activation='relu', name='input_layer'))
```

```python
#Add hidden layer
model.add(Dense(10, kernel_initializer = 'uniform', activation = 'relu', name='hidden_layer_1'))
#Add Output layer
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid', name='output_layer'))
```

```python
#Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
print('Neural Network Model Summary: ')
print(model.summary())
```

```python
# Fitting the ANN
model.fit(X_train,Y_train, epochs=100,validation_data=(X_test,Y_test))
```

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred = (Y_pred > 0.5).astype(np.int)#Predict the results using 0.5 as a threshold
Bank_test_predict=pd.DataFrame(X_test,columns=X_df.columns)
Bank_test_predict["Predicted"]=Y_pred
Bank_test_predict["Actual"]=Bank_df["Exited"]
Bank_test_predict.head()
```

```python
from sklearn import metrics
print(metrics.classification_report(Y_test, Y_pred))
```

```python
print("accuracy", metrics.accuracy_score(Y_test, Y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
```

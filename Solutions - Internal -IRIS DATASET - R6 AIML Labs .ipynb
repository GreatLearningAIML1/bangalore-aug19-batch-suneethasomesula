# Question 3

### Create train and test data
- use train_test_split to get train and test set
- set a random_state
- test_size: 0.25

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=7)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Import `StandardScaler` from `sklearn.preprocessing`
from sklearn.preprocessing import StandardScaler

# Define the scaler 
scaler = StandardScaler().fit(X_train)

# Scale the train set
X_train = scaler.transform(X_train)

# Scale the test set
X_test = scaler.transform(X_test)
```

```python
print(X_train.ndim, X_test.ndim, y_train.ndim, y_test.ndim)
```

## Question 4

### One-hot encode the labels
- convert class vectors (integers) to binary class matrix
- convert labels
- number of classes: 3
- we are doing this to use categorical_crossentropy as loss

```python
#Encoding
import tensorflow as tf

def encode(labels):
    print("Shape of data (BEFORE encode): %s" % str(labels.shape))
    encoded = tf.keras.utils.to_categorical(labels, num_classes = 3, dtype = "float32")
    print("Shape of data (AFTER encode): %s" % str(encoded.shape))
    return encoded
```

```python
encoded_y_train = encode(y_train)
print(encoded_y_train)
```

```python
encoded_y_test = encode(y_test)
print(encoded_y_test)
```
